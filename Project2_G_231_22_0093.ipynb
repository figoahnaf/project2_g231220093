* Membuat Bar Char Tentang Curah Hujan

```

import numpy as np
import pandas as pd
import seaborn as sns
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```
dataset = pd.read_csv('CURAH HUJAN.csv')
dataset
```

```
plt.figure(figsize=(15, 5))

# Menentukan posisi bulan
bulan = np.arange(1, 13)

# Data curah hujan untuk setiap dekade
dset = {
    '1981 - 1990': dataset['1981 - 1990'],
    '1991 - 2000': dataset['1991 - 2000'],
    '2001 - 2010': dataset['2001 - 2010']
}

# Membuat bar chart untuk setiap dekade
colors = ['royalblue', 'firebrick', 'yellowgreen']
labels = ['1981 - 1990', '1991 - 2000', '2001 - 2010']

for i, label in enumerate(labels):
    plt.bar(bulan + 0.2 * i, dataset[label], width=0.2, color=colors[i], label=label)

# Mengatur label sumbu x dan y, serta judul
plt.xlabel('Bulan', fontsize='15', weight='bold')
plt.ylabel('Curah Hujan (mm)',fontsize='15', weight='bold')
plt.xticks(bulan + 0.2, bulan)

# Menambahkan garis-garis horizontal kecil di belakang
for i in range(20, 700, 20):
    plt.axhline(y=i, color='gray', linestyle='-', alpha=0.3)

# Menampilkan plot
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3)
plt.show()
```

* Membuat Line Char Dengan Suhu Permukaan

```
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```
dset = pd.read_csv('SUHU PERMUKAAN.csv')
dset
```

```
plt.figure(figsize=(12, 5))

# Mengambil data CSV
data1 = dset['Asimilasi']
data2 = dset['Observasi']
data3 = dset['Non Asimilasi']
dates = dset['Waktu']

# Membuat line char
plt.plot(dates, data1, 'o-', color='royalblue', label='Asimilasi')
plt.plot(dates, data2, 'o-', color='gold', label='Observasi')
plt.plot(dates, data3, 'o-', color='red', label='Non Asimilasi')


# Buat Judul dan membuat tabel ada di sisi kanan bawah
plt.xlabel('Waktu')
plt.ylabel('Celcius')
plt.title('Suhu Permukaan')
plt.legend(loc='lower right')

for d in range(20, 28):
    plt.axhline(y=d, color='gray', linestyle='-', linewidth=0.3)

plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='-', linewidth=0.3)

plt.show()
```
